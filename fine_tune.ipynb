```python
from dotenv import load_dotenv
from openai import OpenAI
import os

load_dotenv()
open_ai_api_key = os.getenv('OPENAI_API_KEY')
client = OpenAI(
    api_key=open_ai_api_key
)

def upload_file(file_name:str, purpose:str):

    with open(file_name, "rb") as file:
        response = client.files.create(file = file, purpose = purpose)
    return response.id

training_file_id = upload_file("assistant_training.jsonl", "fine-tune")
print("Training file id:")
print(training_file_id)
```

```python

import openai
base_model = "gpt-4.1-nano-2025-04-14"
try:
    response = client.fine_tuning.jobs.create(
        training_file=training_file_id,
        model=base_model,
        suffix="station-messages",
        method = {
            "type": "supervised",
            "supervised": {
                "hyperparameters": {
                    "n_epochs": 1,
                    "learning_rate_multiplier": 0.75
                }
            }
        }
    )
except openai.BadRequestError as e:
    print(f"Bad Request Error: {e.message}")
    print(f"Error Code: {e.code}")
except openai.OpenAIError as e:
    print(f"General OpenAI Error: {e}")

job_id = response.id

print("Job ID:", response.id)
print("Status:", response.status)
```
